La analítica es el proceso de descubrir, interpretar y comunicar patrones significativos en los datos. En pocas palabras, la analítica nos ayuda a ver información y datos significativos que de otro modo no podríamos detectar. La analítica empresarial se centra en el uso de conocimientos derivados de los datos para tomar decisiones más informadas que ayudarán a las organizaciones a aumentar las ventas, reducir los costos y realizar otras mejoras comerciales.

La analítica prescriptiva es la rama de la analítica de datos que se encarga en responder las preguntas: ¿Qué debemos hacer? ¿Cuál es el mejor curso de acción? y por lo tanto es diferente de la analítica predictiva, que responde preguntas del tipo: ¿Qué pasará?. El valor de la analítica de datos proporciona fuentes de conocimiento que pueden ahorrar millones de dólares en operaciones y maximizar la eficiencia de las empresas. Por lo tanto, implementar tecnologías de toma de decisiones es crucial para que un negocio prospere.

El siguiente modelo de optimización es una herramienta matemática Last-Interaction que busca encontrar la mejor manera posible de planear un presupuesto publicitario mensual en distintos medios de comunicación para una amplia familia de productos. El modelo optimiza el valor esperado de los KPI’s de negocio más relevantes y sugiere cursos de acción específicos para el alcance de los mismos bajo el cumplimiento de ciertas reglas y restricciones de negocio. Este es un modelo analítico prescriptivo por lo que no debe confundirse con un modelo analítico predictivo ya que su naturaleza en el modelamiento difiere significativamente.

### Conjuntos, subconjuntos e índices

* $I$: Conjunto de productos, $i \in {1,2...,I}$
* $C \subset I$: Subconjunto de productos que representan conversiones, $c \in {1,2,...,C}$
* $L \subset I$: Subconjunto de productos que representan leads, $l \in {1,2,...,L}$
* $J$: Conjunto de medios de comunicación, $j \in {1,2...,J}$

### Parámetros (datos)
Las tasas de conversión, leads y de clicks están influenciadas por el efecto que tiene las impresiones en cada medio. Su calculo depende escencialmente de los datos históricos y la calidad de los mismos.

* $\mu_{i,j}^{conv}$: Tasa de conversiones esperadas por unidad de inversión para el producto ${i}$ en el medio de comunicación ${j}$, $\forall i \in C \subset I, \forall j \in J$
$$ \mu_{i,j}^{conv} = \frac{1}{T}\sum_{t = 1}^{T} \frac{a_{i,j,t}}{b_{i,j,t}} * \frac{1}{T}\sum_{t = 1}^{T} \frac{c_{i,j,t}}{a_{i,j,t}},  \forall i \in C \subset I, \forall j \in J $$

* $\mu_{i,j}^{leads}$: Tasa de leads esperadas por unidad de inversión para el producto ${i}$ en el medio de comunicación ${j}$, $\forall i \in L \subset I, \forall j \in J$
$$ \mu_{i,j}^{leads} = \frac{1}{T}\sum_{t = 1}^{T} \frac{a_{i,j,t}}{b_{i,j,t}} * \frac{1}{T}\sum_{t = 1}^{T} \frac{d_{i,j,t}}{a_{i,j,t}},  \forall i \in L \subset I, \forall j \in J $$

* $\gamma_{i,j}$: Tasa de clicks esperadas por unidad de inversión para el producto ${i}$ en el medio de comunicación ${j}$, $\forall i \in I, \forall j \in J$
$$ \gamma_{i,j} = \frac{1}{T}\sum_{t = 1}^{T} \frac{a_{i,j,t}}{b_{i,j,t}} * \frac{1}{T}\sum_{t = 1}^{T} \frac{e_{i,j,t}}{a_{i,j,t}},  \forall i \in  I, \forall j \in J $$

* $\delta_{i,j}^{cpa}$: Costo promedio por adquisición esperado para el producto ${i}$ en el medio de comunicación ${j}$, $\forall i \in C \subset I, \forall j \in J$
$$ \delta_{i,j}^{cpa} = \frac{1}{T}\sum_{t = 1}^{T} \frac{b_{i,j,t}}{c_{i,j,t}},  \forall i \in C \subset I, \forall j \in J $$

* $\delta_{i,j}^{cpl}$: Costo promedio por lead esperado para el producto ${i}$ en el medio de comunicación ${j}$, $\forall i \in L \subset I, \forall j \in J$
$$ \delta_{i,j}^{cpl} = \frac{1}{T}\sum_{t = 1}^{T} \frac{b_{i,j,t}}{d_{i,j,t}},  \forall i \in L \subset I, \forall j \in J$$

* $\beta_{j}^{inf}$: Inversión mínima en el medio de comunicación ${j}$, $\forall j \in J$
* $\beta_{j}^{sup}$: Límite máximo de inversión en el medio de comunicación ${j}$, $\forall j \in J$
* $\omega_{i}^{min}$: Importancia mínima porcentual relativa del producto ${i}$, $\forall i \in I$
* $\omega_{i}^{max}$: Asignación porcentual máxima del producto ${i}$, $\forall i \in I$
* $\sigma_{i}^{2}$: Nivel de riesgo aceptable para el producto ${i}$, $\forall i \in C \subset I$
* $\lambda_{i}$: Presupuesto idealmente destinado para el producto ${i}$, $\forall i \in I$
* $\nu_{i}$: Totalidad de resultado combinado esperado (conversiones/leads) para el producto ${i}$, $\forall i \in I$
* $c_{i}^{cpa}$: Costo promedio por adquisición esperado (a total) para el producto ${i}$, $\forall i \in I$
* $\sigma_{total}^{2}$: Nivel de riesgo aceptable general (totalidad para medios).
* $\Beta$: Presupuesto total para las campañas publicitarias.
* $\Gamma$: Indicador mínimo requerido para el número de clicks totales.
* $\Mu$: Indicador mínimo requerido para el número de conversiones totales.
* $\Pi$: Indicador mínimo requerido para el número de leads totales.
* $\Nu$: Indicador mínimo requerido para el resultado combinado total (conversiones/leads).
* $\Delta$: Indicador máximo permitido para el costo promedio por adquisición general.
* $C$: Indicador máximo permitido para el costo promedio por adquisición general (productos).
* $\Omicron$: Indicador máximo permitido para el costo promedio por lead general.
* $\Sigma_{j,k,i}$: Covarianza entre las conversiones/leads históricas del medio de comunicación ${j}$ y el medio de comunicación ${k}$ para el producto ${i}$, $\forall i \in I, \forall (j,k) \in J$
* $\Sigma_{i,p}^{total}$: Covarianza entre las conversiones/leads históricas del producto ${i}$ y el producto ${p}$ (totalidad en medios), $\forall (i,p) \in I$


### Variables de decisión
* $x_{i,j}$: La cantidad de dinero del presupuesto publicitario asignado al producto ${i}$ en el medio de comunicación ${j}$, $\forall i \in I, \forall j \in J$
* $y_{i}$: La proporción del presupuesto publicitario asignado únicamente al producto ${i}$ (totalidad en cada medio), $\forall i \in I$
* $z_{i,p}$: Variable auxiliar definida para linealizar la primera restricción cuadrática: $z_{i,p}$ = $y_{i}*y_{p}$, $\forall (i,p) \in I$
* $u_{i,j,k}$: Variable auxiliar definida para linealizar la segunda restricción cuadrática: $u_{i,j,k}$ = $x_{i,j}*x_{i,k}$, $\forall (i,p) \in I$

### Funciones objetivo

   * **Maximizar las conversiones esperadas totales.**

 $$ \ Max\space \ Z_1 = \sum_{i \in C \cup L \subset I}\sum_{j \in J} \mu_{i,j}^{conv}*x_{i,j} $$

   * **Maximizar los leads esperados totales.**

 $$ \ Max\space \ Z_2 = \sum_{i \in L \cup C \subset I}\sum_{j \in J} \mu_{i,j}^{leads}*x_{i,j} $$

  * **Maximizar el número de clicks esperados totales**.

 $$ \ Max\space \ Z_3 = \sum_{i \in I}\sum_{j \in J} \gamma_{i,j}*x_{i,j} $$

  * **Minimizar el costo por adquisición esperado total**.

 $$ \ Min\space \ Z_4 = \sum_{i \in C \cup L \subset I}\sum_{j \in J} \delta_{i,j}^{cpa}*\frac{x_{i,j}}{\Beta} $$

   * **Minimizar el costo por lead esperado total**.

 $$ \ Min\space \ Z_5 = \sum_{i \in L \cup C \subset I}\sum_{j \in J} \delta_{i,j}^{cpl}*\frac{x_{i,j}}{\Beta} $$

  * **Minimizar el riesgo de inversión para todos los productos**.

 $$ \ Min\space \ Z_6 = \sum_{i \in I}\sum_{p \in I} y_{i}*y_{p}*\Sigma_{i,p}^{total}$$

   * **Maximizar el resultado combinado total (conversiones/leads)**.

 $$ \ Max\space \ Z_7 = \sum_{i \in I} \nu_{i} * y_{i} $$

   * **Minimizar el costo por adquisición esperado total (productos)**.

 $$ \ Min\space \ Z_8 = \sum_{i \in I} c_{i}^{cpa}*y_{i} $$

 ### Restricciones

 **1. Satisfacción del presupuesto publicitario**.

 Esta restricción asegura que el presupuesto total asignado a las campañas publicitarias se satisface si llega a ser necesario.

 $$ \sum_{i \in I} \sum_{j \in J} x_{i,j} = \Beta $$
 $$ \Beta = \sum_{i \in I} \lambda_{i} $$

 Cuando únicamente se tiene en cuenta la asignación del presupuesto para cada producto teniendo encuenta la totalidad de medios de comunicación, la suma de las ponderaciones debe ser igual a uno.

 $$ \sum_{i \in I} y_{i} = 1 $$

 **2. Límites de inversión para cada medio de comunicación**.

 Estas restricciones establecen los límites mínimos ($\beta_{j}^{inf}$) y máximos ($\beta_{j}^{sup}$) de inversión en cada medio de comunicación (${j}$). Si se cumplen estos límites, la restricción es factible.

 $$ \sum_{i \in I} x_{i,j} \geq \beta_{j}^{inf},  \forall j \in J $$
 $$ \sum_{i \in I} x_{i,j} \leq \beta_{j}^{sup},  \forall j \in J $$

 **3. Importancia mínima y asignación máxima por producto**.

 Esta restricción garantiza que la inversión en cada producto (${i}$) cumpla con un nivel mínimo de importancia porcentual relativa ($\omega_{i}^{min}$) y también esté sujeto a un máximo de asignación porcentual en cada producto ($\omega_{i}^{max}$). Si esta condición se satisface para todos los productos, la restricción es factible.

 $$ \sum_{j \in J} \frac{x_{i,j}}{\Beta} \geq \omega_{i}^{min}, \forall i \in I $$
 $$ \sum_{j \in J} \frac{x_{i,j}}{\Beta} \leq \omega_{i}^{max}, \forall i \in I $$

 $$ y_{i} \geq \omega_{i}^{min}, \forall i \in I $$
 $$ y_{i} \leq \omega_{i}^{max}, \forall i \in I $$

 **4. Satisfacción para el indicador mínimo de clicks**.

 Esta restricción exige que el número total de clicks esperados sea mayor o igual a un valor mínimo requerido ($\Gamma$). Si esta condición se cumple, la restricción es factible.

 $$ \sum_{i \in I}\sum_{j \in J} \gamma_{i,j}*x_{i,j} \geq \Gamma $$

 **5. Satisfacción para el indicador mínimo de conversiones y leads**.

 Esta restricción establece que el número total de conversiones y leads esperados debe ser mayor o igual a un valor mínimo requerido ($\Mu$ y $\Pi$). Si esta condición se satisface, la restricción es factible.

 $$ \sum_{i \in C \cup L \subset I}\sum_{j \in J} \mu_{i,j}^{conv}*x_{i,j} \geq \Mu $$
 $$ \sum_{i \in L \cup C \subset I}\sum_{j \in J} \mu_{i,j}^{leads}*x_{i,j} \geq \Pi $$
 $$ \sum_{i \in I} \nu_{i} * y_{i} \geq \Nu$$

 **6. Límite para el indicador máximo permitido del costo promedio por adquisición general y el costo promedio por leads general**.

 Esta restricción impone un límite superior para el costo promedio por adquisición general y el costo promedio por lead general ($\Delta$). Si esta condición se cumple, la restricción es factible.

 $$ \sum_{i \in C \cup L \subset I}\sum_{j \in J} \delta_{i,j}^{cpa}*\frac{x_{i,j}}{\Beta} \leq \Delta $$
 $$ \sum_{i \in L \cup C \subset I}\sum_{j \in J} \delta_{i,j}^{cpl}*\frac{x_{i,j}}{\Beta} \leq \Omicron $$

 $$ \delta_{i,j}^{cpa} \neq \frac{1}{\mu_{i,j}^{conv}}, \forall i \in I, \forall j \in J $$
 $$ \delta_{i,j}^{cpl} \neq \frac{1}{\mu_{i,j}^{leads}}, \forall i \in I, \forall j \in J $$

 $$ \sum_{i \in I} c_{i}^{cpa}*y_{i} \leq C $$

 **7. Diversificación del riesgo para la inversión publicitaria en cada producto**.

 Esta restricción establece un límite para la diversificación del riesgo en la inversión publicitaria para cada producto ($i$). Si esta condición se satisface para todos los productos, la restricción es factible.

 $$ \sum_{j \in J}\sum_{k \in J} \frac{x_{i,j}*x_{i,k}*\Sigma_{j,k,i}}{\Beta} \leq \sigma_{i}^{2}, \forall i \in I $$

 **8. Diversificación del riesgo para la inversión publicitaria en totos los productos**.

 Esta restricción establece un límite para la diversificación del riesgo en la inversión publicitaria de forma general. Si esta condición se satisface para todos los productos, la restricción es factible. 

 $$ \sum_{i \in I}\sum_{p \in I} y_{i}*y_{p}*\Sigma_{i,p}^{total} \leq \sigma_{total}^{2} $$

 **9. Linealización de la restricción cuadrática**.

 Esta restricción tiene en cuenta la variable auxiliar para linealizar la restricción cuadrática de la diversificación del riesgo usando el método de McCormick. Si esta condición se cumple para todos los productos, la restricción es factible.

 #### Primera restricción

 $$ \sum_{i \in I}\sum_{p \in I} z_{i,p}*\Sigma_{i,p}^{total} \leq \sigma_{total}^{2} $$

 $$ z_{i,p} \geq y_{i}^{L}y_{p} + y_{i}y_{p}^{L} - y_{i}^{L}y_{p}^{L}, \forall(i,p) \in I $$

 $$ z_{i,p} \geq y_{i}^{U}y_{p} + y_{i}y_{p}^{U} - y_{i}^{U}y_{p}^{U}, \forall(i,p) \in I $$
 
 $$ z_{i,p} \leq y_{i}^{U}y_{p} + y_{i}y_{p}^{L} - y_{i}^{U}y_{p}^{L}, \forall(i,p) \in I $$

 $$ z_{i,p} \leq y_{i}y_{p}^{U} + y_{i}^{L}y_{p} - y_{i}^{L}y_{p}^{U}, \forall(i,p) \in I $$


 #### Segunda restricción

 $$ \sum_{j \in J}\sum_{k \in J} \frac{u_{i,j,k}*\Sigma_{j,k,i}}{\Beta} \leq \sigma_{i}^{2}, \forall i \in I $$

 $$ u_{i,j,k} \geq x_{i,j}^{L}x_{i,k} + x_{i,j}x_{i,k}^{L} - x_{i,j}^{L}x_{i,k}^{L}, \forall (j,k) \in J, \forall i \in I  $$

 $$ u_{i,j,k} \geq x_{i,j}^{U}x_{i,k} + x_{i,j}x_{i,k}^{U} - x_{i,j}^{U}x_{i,k}^{U}, \forall (j,k) \in J, \forall i \in I  $$

 $$ u_{i,j,k} \leq x_{i,j}^{U}x_{i,k} + x_{i,j}x_{i,k}^{L} - x_{i,j}^{U}x_{i,k}^{L}, \forall (j,k) \in J, \forall i \in I  $$

 $$ u_{i,j,k} \leq x_{i,j}x_{i,k}^{U} + x_{i,j}^{L}x_{i,k} - x_{i,j}^{L}x_{i,k}^{U}, \forall (j,k) \in J, \forall i \in I  $$

 **10. No negatividad**.

 Esta restricción establece que la cantidad de dinero presupuestada en cada producto y medio de comunicación debe ser no negativa. Por otra parte, la variable $y_{i}$ debe estar entre cero y uno.
 
$$ x_{i,j} \geq 0, \forall i \in I, \forall j \in J $$
$$ 0 \leq y_{i} \leq 1, \forall i \in I $$
$$ z_{i,p} \geq 0, \forall(i,p) \in I $$
$$ u_{i,j,k} \geq 0, \forall (j,k) \in J, \forall i \in I$$
